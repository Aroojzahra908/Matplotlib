# stacked bar chart

# percentage,,,,autopct='%0.1f%%'

```python
name = ['Aa', 'Bb', 'Cc', 'Dd', 'Ee', 'Ff', 'Gg', 'Hh', 'Ii']
marks = [10, 23, 78, 95, 38, 37, 77, 48, 33]

D = plt.pie(marks, labels=name, autopct='%0.1f%%')
plt.title('Marks Distribution')
plt.show()
```

# fill area between curves

```python

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='Curve 1')
plt.plot(x, y2, label='Curve 2')
plt.fill_between(x, y1, y2, color='blue', alpha=0.3)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Filled Area Example')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='Curve 1')
plt.plot(x, y2, label='Curve 2')
plt.fill_between(x, y1, y2, color='blue', alpha=0.3)
plt.fill_between(x, y1, y2, where=(y1 < y2), color='red', alpha=0.6)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Filled Area Example')
plt.legend()
plt.show()
```

# subplot


```python
fig, axis = plt.subplots(2, 2)
axis[0, 0].plot([1, 2, 3, 5], [3, 4, 5, 6])      
axis[0,0].title.set_text('0 row, 0 column')


axis[0, 1].plot([1, 2, 3, 8], [3, 4, 9, 6])      
axis[0,1].title.set_text('0 row, 1 column')


axis[1, 1].plot([1, 2, 6, 8], [10, 4, 9, 6])      
axis[1,1].title.set_text('1 row, 1 column')

# Adjust spacing between subplots
plt.tight_layout()
plt.show()
```
